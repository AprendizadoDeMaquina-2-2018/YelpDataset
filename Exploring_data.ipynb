```python
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
yelp_100 = pd.read_csv('dataset/yelp_academic_dataset_review_reduced_100.csv')
yelp_100.head()
```

```python
reviews = yelp_100['text'].values
print(reviews[0])
review_ids = yelp_100['review_id'].values
print(review_ids[0])
```

```python
tfidf = TfidfVectorizer()
csr_mat = tfidf.fit_transform(reviews)
print(csr_mat)
print(csr_mat.toarray())
```

```python
words = tfidf.get_feature_names()
words
```

```python
# Create a TruncatedSVD instance: svd
svd = TruncatedSVD(n_components=50)

# Create a KMeans instance: kmeans
kmeans = KMeans(n_clusters=2)

# Create a pipeline: pipeline
pipeline = make_pipeline(svd, kmeans)
```

```python
# Fit the pipeline to articles
pipeline.fit(csr_mat)

# Calculate the cluster labels: labels
labels = pipeline.predict(csr_mat)

# Create a DataFrame aligning labels and titles: df
df = pd.DataFrame({'label': labels, 'texts': reviews})

# Display df sorted by cluster label
print(df.sort_values('label'))
```
